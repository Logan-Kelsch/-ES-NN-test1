```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate a custom random value with skewed distributions
# toward the center (around 50), while keeping peaks near 10 and 90
def custom_random_skewed():
    rand_val = np.random.rand()
    if rand_val < 0.3:
        return np.random.normal(loc=15, scale=6)  # slightly closer to 50 than 10
    elif rand_val < 0.6:
        return np.random.normal(loc=85, scale=6)  # slightly closer to 50 than 90
    else:
        return np.random.uniform(0, 100)  # background noise

# Clamp values to the range [0, 100]
def bounded_random_skewed():
    val = custom_random_skewed()
    return np.clip(val, 0, 100)

# Generate samples
samples_skewed = [bounded_random_skewed() for _ in range(10000)]

# Plot the histogram
plt.figure(figsize=(10, 5))
plt.hist(samples_skewed, bins=100, density=True, color='steelblue', edgecolor='black')
plt.title("Custom Distribution with Skewed Peaks Toward Center")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.tight_layout()
plt.show()
```
